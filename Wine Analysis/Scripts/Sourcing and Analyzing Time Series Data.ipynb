### Testing for stationarity

```python
from statsmodels.tsa.stattools import adfuller

def dickey_fuller(timeseries):
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

dickey_fuller(data['Value'])
```

Test statistic is larger than the critical value, so we must make the data stationary.

```python
# Check out a plot of autocorrelations

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(data)
plt.show()
```

The plot supports the result of the Dickey-Fuller test, showing that the data requires stationarizing.

### Stationarize the data

```python
# Stationarizing the data using the df.shift(1) function

data_diff = data - data.shift(1)
```

```python
# Remove missing values that came about due to differencing

data_diff.dropna(inplace = True)
```

```python
# Check headers

data_diff.head()
```

```python
# Check columns

data_diff.columns
```

```python
# Check out what the differencing did to the time-series curve

plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff)
```

```python
# Perform Dickey-Fuller test again

dickey_fuller(data_diff)
```

```python
# Check autocorrelations

plot_acf(data_diff)
plt.show()
```

Test statistic is now smaller than the critical value, and there are only two lines above the blue confidence interval. I can consider this series stationary.
